# การรับค่า input() และแสดงผลด้วย print()

ในบทเรียนนี้เราจะเรียนรู้เกี่ยวกับฟังก์ชัน `input()` และ `print()` ซึ่งเป็นฟังก์ชันพื้นฐานที่สำคัญในการเขียนโปรแกรมภาษา Python

## ฟังก์ชัน print()

`print()` คือฟังก์ชันที่ใช้ในการแสดงผลข้อมูลออกทางหน้าจอ

### รูปแบบการใช้งาน
```python
print(objects, sep=' ', end='\n')"
```

- `objects`: ข้อมูลที่ต้องการแสดงผล (สามารถใส่ได้หลายตัว คั่นด้วยเครื่องหมาย `,`)
- `sep`: ตัวคั่นระหว่างข้อมูล (ค่าเริ่มต้นคือช่องว่าง)
- `end`: ตัวอักษรที่จะแสดงหลังจากแสดงผลเสร็จ (ค่าเริ่มต้นคือขึ้นบรรทัดใหม่ `\n`)

````python
# ตัวอย่างการใช้งาน print() แบบง่าย
print("Hello World!")
print("ยินดีต้อนรับสู่คอร์ส Python")
````

````python
# การแสดงผลหลายค่าใน print() เดียว
name = "Python"
version = 3.10

print("ภาษา", end="-")
print(name, end="-")
print("เวอร์ชัน", end="-")
print(version)

print("ภาษา", name, "เวอร์ชัน", version)
````

````python
# การกำหนด sep และ end
print("วันจันทร์", "วันอังคาร", "วันพุธ", sep="-")
print("สวัสดี", "ทุกคนที่รับชม", end="!!!\n")
print("ทุกคน")
````

##  Escape Sequences

````python
# ตัวอย่างการใช้ Escape Sequences
print("Hello \nWorld")      # \n ใช้สำหรับขึ้นบรรทัดใหม่
print("\'Double Quote\'")   # \' ใช้สำหรับแสดงเครื่องหมายคำพูด
print("\"Double Quote\"")   # \" ใช้สำหรับแสดงเครื่องหมายคำพูด
print("ใช้แบ็คสแลช \\")       # \\ ใช้สำหรับแสดงเครื่องหมายแบ็คสแลช
print("กดเว้นววรรค \tTab")   # \t ใช้สำหรับเพิ่มแท็บ
````

## ฟังก์ชัน input()

`input()` คือฟังก์ชันที่ใช้รับข้อมูลจากผู้ใช้ผ่านทางคีย์บอร์ด

### รูปแบบการใช้งาน
```python
input(prompt)
```

- `prompt`: ข้อความที่จะแสดงเพื่อบอกผู้ใช้ว่าต้องการให้กรอกข้อมูลอะไร

### สิ่งที่ควรรู้
- ค่าที่ได้จาก `input()` จะเป็น **string** (ข้อความ) เสมอ
- หากต้องการรับค่าตัวเลข ต้องแปลงชนิดข้อมูลด้วย `int()` หรือ `float()`

````python
input_str = input("กรุณาใส่ตัวเลข 2 ตัว(เว้นวรรคคั่น): ")
print(input_str)
print(input_str.split()[0])
print(input_str.split()[1])
````

````python
# ตัวอย่างการใช้งาน input() แบบง่าย
fullname = input("กรุณาใส่ชื่อและนามสกุล(เว้นวรรคคั่น): ")
name = fullname.split()[0]
surname = fullname.split()[1]
# แสดงผลชื่อที่กรอกเข้ามา
print("สวัสดีคุณ", name, surname)
````

## ตัวอย่างที่ 1: โปรแกรมคำนวณพื้นที่สี่เหลี่ยม

โปรแกรมนี้จะรับค่าความกว้างและความยาวของสี่เหลี่ยมจากผู้ใช้ แล้วคำนวณพื้นที่  
Input:  
5  
4  

Output:
===== โปรแกรมคำนวณพื้นที่สี่เหลี่ยม =====  
พื้นที่สี่เหลี่ยม = 20.0 ตารางเมตร  
พื้นที่สี่เหลี่ยมขนาด 5.0 x 4.0 เมตร = 20.0 ตารางเมตร

````python
# โปรแกรมคำนวณพื้นที่สี่เหลี่ยม
print("===== โปรแกรมคำนวณพื้นที่สี่เหลี่ยม =====")

# รับค่าความกว้างและความยาว
width = float(input("ความกว้าง (เมตร): "))
length = float(input("ความยาว (เมตร): "))

# คำนวณพื้นที่
area = width * length

# แสดงผลลัพธ์
print("พื้นที่สี่เหลี่ยม =", area, "ตารางเมตร")
print(f"พื้นที่สี่เหลี่ยมขนาด {width} x {length} เมตร = {area} ตารางเมตร")
````

## ตัวอย่างที่ 2: โปรแกรมทักทายตามช่วงเวลา

โปรแกรมนี้จะรับชื่อและเวลาปัจจุบัน (ชั่วโมง) จากผู้ใช้ แล้วแสดงคำทักทายที่เหมาะสมกับช่วงเวลา

Input:  
LouisMakerLab  
15  

Output:  
===== โปรแกรมทักทายตามช่วงเวลา =====  
สวัสดีตอนบ่าย คุณlouisMakerLab

````python
# โปรแกรมทักทายตามช่วงเวลา
print("===== โปรแกรมทักทายตามช่วงเวลา =====")

# รับชื่อและเวลาจากผู้ใช้
name = input("กรุณาใส่ชื่อของคุณ: ")
hour = int(input("ขณะนี้เวลากี่โมง (0-23): "))

# กำหนดคำทักทายตามช่วงเวลา
output = ""
if 5 <= hour < 12:
    output = "อรุณสวัสดิ์"
elif 12 <= hour < 17:
    output = "สวัสดีตอนบ่าย"
elif 17 <= hour < 21:
    output = "สวัสดีตอนเย็น"
else:
    output = "สวัสดีตอนกลางคืน"

# แสดงผลการทักทาย
print(f"{output} คุณ{name}")
````

## ตัวอย่างที่ 3: โปรแกรมแนะนำตัวเอง

โปรแกรมนี้จะรับข้อมูลส่วนตัวจากผู้ใช้ แล้วนำมาแสดงผลเป็นข้อความแนะนำตัวเอง

Input:  
LouisMakerLab  
25  
Youtuber  
code  

Output:  
===== โปรแกรมแนะนำตัวเอง =====  
สวัสดีครับ/ค่ะ ฉันชื่อ LouisMakerLab อายุ 25 ปี  
ฉันทำงานเป็น youtuber และมีงานอดิเรกคือ code  
ยินดีที่ได้รู้จักทุกคนครับ/ค่ะ  

````python
# โปรแกรมแนะนำตัวเอง
print("===== โปรแกรมแนะนำตัวเอง =====")

# รับข้อมูลจากผู้ใช้
name = input("ชื่อ: ")
age = input("อายุ: ")
job = input("อาชีพ: ")
hobby = input("งานอดิเรก: ")

# แสดงผลข้อความแนะนำตัว
print(f"สวัสดีครับ/ค่ะ ฉันชื่อ {name} อายุ {age} ปี")
print(f"ฉันทำงานเป็น {job} และมีงานอดิเรกคือ {hobby}")
print("ยินดีที่ได้รู้จักทุกคนครับ/ค่ะ")
````

## สรุปความรู้

### ฟังก์ชัน print()
- ใช้สำหรับแสดงผลข้อมูลออกทางหน้าจอ
- สามารถแสดงผลได้ทั้งข้อความ, ตัวเลข, และตัวแปร
- สามารถกำหนดตัวคั่น (`sep`) และตัวสิ้นสุด (`end`) ได้

### ฟังก์ชัน input()
- ใช้รับข้อมูลจากผู้ใช้ผ่านทางคีย์บอร์ด
- ค่าที่รับมาจะเป็นชนิด `str` (string) เสมอ
- ต้องแปลงชนิดข้อมูลด้วย `int()`, `float()` หากต้องการใช้งานเป็นตัวเลข

### การนำไปใช้งาน
- ใช้ในการสร้างโปรแกรมที่มีการโต้ตอบกับผู้ใช้
- การรับข้อมูลและแสดงผลเป็นพื้นฐานสำคัญในการพัฒนาโปรแกรมทุกประเภท

## แบบฝึกหัด

2. เขียนโปรแกรมคำนวณค่า BMI โดยรับน้ำหนัก (กิโลกรัม) และส่วนสูง (เซนติเมตร) จากผู้ใช้ สูตร BMI = น้ำหนัก / (ส่วนสูง/100)²

3. เขียนโปรแกรมแปลงอุณหภูมิระหว่าง Celsius และ Fahrenheit โดยรับค่าอุณหภูมิและหน่วยจากผู้ใช้
